```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import statistics
import matplotlib.ticker as mticker
import plotly.graph_objects as go
```

```python
# extract all csvs and put in 1 folder
for folder in os.listdir("../Companies"):
    if os.path.isdir(f"../Companies/{folder}"):
        for csv_file in os.listdir(f"../Companies/{folder}"):
            print(csv_file)
            os.rename(f"../Companies/{folder}/{csv_file}", f"../Companies/{csv_file}")
        os.rmdir(f"../Companies/{folder}")
```

# Removing csvs with < 60 days (2 months worth of data)
https://smartasset.com/investing/stock-correlation -> Stock correlation is seen over months or years,
rather than days to see if the correlation between the pairs are truly consistent
rather than a short term event as with long term correlation, we are able to exploit inconsistencies in
stock prices

```python
lengths_of_data = []
all_csvs = os.listdir("../Companies")
for csv_file in all_csvs:
    df = pd.read_csv(f"../Companies/{csv_file}")
    lengths_of_data.append(df.shape[0])
```

```python
print("largest csv: ", max(lengths_of_data))
print("smallest csv: ", min(lengths_of_data))
print("Average number of rows in csvs: ", sum(lengths_of_data)/len(lengths_of_data))
print("Median number of rows in csvs: ", statistics.median(lengths_of_data))
```

```python
# Due to the large difference in csv sizes, we see that we need to remove csv files with fewer than 60 rows
# as that is not enough data to provide a true correlation

for csv_file in all_csvs:
    df = pd.read_csv(f"../Companies/{csv_file}")
    if df.shape[0] < 60:
        print(f"Removed file {csv_file} -> {df.shape[0]} rows")
        os.remove(f"../Companies/{csv_file}")
```

# Removing the rows with NaN values in all the csv files

```python
for csv in os.listdir("../Companies/"):
    df = pd.read_csv(f"../Companies/{csv}")
    df.dropna()
    os.remove(f"../Companies/{csv}")
    df.to_csv(f"../Companies/{csv}")
```
